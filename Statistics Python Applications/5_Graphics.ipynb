```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset("mpg")
```

About Dataset: The mpg dataset is included in the Seaborn library and contains information related to the fuel efficiency (miles per gallon, MPG) of cars. This dataset includes various features of different car models.

    mpg: The fuel efficiency of the car (miles per gallon).
    cylinders: The number of cylinders in the engine.
    displacement: Engine displacement (in cubic inches).
    horsepower: The power of the engine (horsepower).
    weight: The weight of the car (in pounds).
    acceleration: The time it takes to accelerate from 0 to 60 miles per hour (in seconds).
    model_year: The model year of the car.
    origin: The region where the car was manufactured (1: USA, 2: Europe, 3: Japan).
    name: The name of the car.


    mpg: Otomobilin yakıt verimliliği (mil başına galon).
    cylinders: Motorun silindir sayısı.
    displacement: Motor hacmi (kübik inç cinsinden).
    horsepower: Motor gücü (beygir gücü).
    weight: Otomobilin ağırlığı (pound cinsinden).
    acceleration: 0'dan 60 mil/sa hıza ulaşma süresi (saniye cinsinden).
    model_year: Otomobilin model yılı.
    origin: Otomobilin üretildiği bölge (1: Amerika, 2: Avrupa, 3: Japonya).
    name: Otomobilin adı.

```python
sns.countplot(x="origin", data=df);#sns.countplot kategorik bir değişkenin değerlerinin sayısını bar chart ile görselleştirmek için kullanılır. 
print("Bu grafik mpg veri setindeki araçların origin sütununa göre dağılımını görsel olarak anlamanızı sağlar ve veri setinin bölgesel çeşitliliği hakkında hızlı bir genel bakış sunar.")
```

```python
labels = df.origin.unique() # origin sütununda bulunan benzersiz değerleri alır.

sizes = df.origin.value_counts()  #origin sütunundaki her bir benzersiz değerin sayısını alır.

explode = (0, 0, 0.1)  # only "explode" the 2nd slice (i.e. 'Hogs') pasta dilimlerinden birini öne çıkarmak için kullanılır.

fig1, ax1 = plt.subplots(figsize=(6,6)) # 6x6 boyutlarında bir grafik figürü ve eksen nesnesi oluşturur.

ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',  # autopct='%1.1f%%' her dilimin yüzdesini gösterir
        shadow=True, startangle=90) # startangle=90 ilk dilimin üstten başlaması için grafiği grafiği 90 derece döndürür.

ax1.axis('equal')  # Equal pasta grafiğinin bir daire olarak çizilmesini sağlar.

plt.show()
```
